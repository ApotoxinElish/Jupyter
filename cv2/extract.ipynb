```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
video = cv2.VideoCapture("./loki.mp4")
num = 0  # 计数器
save_step = 300  # 间隔帧
while True:
    ret, frame = video.read()
    if not ret:
        break
    num += 1
    if num % save_step == 0:
        cv2.imwrite("./demo_images/" + str(num) + ".jpg", frame)
```
