```python
import sympy as sp
import numpy as np
```

```python
# Example 3.1
A = sp.Matrix([[2, 3], [3, -6]])
A
```

```python
# Characteristic polynomial
A.charpoly()
```

```python
# Factorize the polynomial
sp.factor(A.charpoly().as_expr())
```

```python
# Eigenvalues of A
A.eigenvals()
```

```python
# Eigenvalues and eigenvectors of A
A.eigenvects()
```

```python
# Example 3.2
B = sp.Matrix([[4, -1, 6], [2, 1, 6], [2, -1, 8]])
B
```

```python
# Eigenvalues and eigenvectors of B
B.eigenvects()
```

```python
# Example 3.3
C = sp.Matrix([[4, 0, -1, -2], [-1, 3, 0, 1], [0, 0, 4, 0], [1, 1, -1, 1]])
C
```

```python
# Eigenvalues and eigenvectors of C
C.eigenvects()
```

```python
# Example 3.2
B = np.array([[4, -1, 6], [2, 1, 6], [2, -1, 8]])
B
```

```python
# Eigenvalues and eigenvectors of B @ NumPy
np.linalg.eig(B)
```

```python
# A 10x10 random matrix
R = np.random.randint(0, 4, size=(10, 10))
R = R + R.T
R
```

```python
# Eigenvalues and eigenvectors of R @ NumPy
np.linalg.eig(R)
```

```python
# Same matrix in SymPy
S = sp.Matrix(R)
S
```

```python
# Eigenvalues and eigenvectors of R @ SymPy
S.eigenvects()
```
