```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


# define the function f(t,X)
def f(t, X):
    x = X[0]
    y = X[1]
    dx = 2 * x + 2 * y
    dy = x + 3 * y
    dX = np.array([dx, dy])
    return dX


# Solve the ODE numerically
sol = solve_ivp(f, [0, 1], np.array([-1, 2]), t_eval=np.arange(0, 1, 0.05))
# Plot the solution
plt.figure(figsize=(6, 6))
for y in sol.y:
    plt.plot(sol.t, y, label=str(y))
plt.xlabel("t")
plt.ylabel("X(t)")
plt.show()
plt.figure(figsize=(6, 6))
plt.plot(sol.y[0], sol.y[1])
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.show()
```
