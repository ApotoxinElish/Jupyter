**QUESTION 2**

An Ada-Boost algorithm based on parallel axis weak classifiers is used to build a strong classifier.

There are 400 samples, in the training data set.

 Initially, at step time t=1 all normalized weights Dt (t=1) are equal, a weak classifier hs( ) which gives the least error is selected. Using hs( ), 351 samples are correctly classified.  

At time t=2, the normalized weight for a correctly classified sample is Dt_correct (t=2) (Assume all normalized weights for correctly classified samples are equal at this stage.)

Find Dt_correct (t=2).

**Answer:0.00142**

````python
import math
num_instance = 400
D_t1 = 1/num_instance
num_correct = 351
num_incorrect = num_instance - num_correct

err_t1 = D_t1 * 0 * num_correct + D_t1 * 1 * num_incorrect
alpha_t1 = (1/2) * math.log((1-err_t1)/err_t1)
# for correct instance: yi_hi = 1; for incorrect instance: yi_hi = -1;
Z_t1 = D_t1 * math.pow(math.e, -1 * alpha_t1 * 1) * num_correct + D_t1 * math.pow(math.e, -1 * alpha_t1 * -1) * num_incorrect

D_t2_correct = (D_t1 * math.pow(math.e, -1 * alpha_t1 * 1)) / Z_t1
D_t2_incorrect = (D_t1 * math.pow(math.e, -1 * alpha_t1 * -1)) / Z_t1

print(f'D_t1={D_t1}')
print(f'err_t1={err_t1}')
print(f'alpha_t1={alpha_t1}')
print(f'Z_t1={Z_t1}')
print(f'D_t2_correct={D_t2_correct}')
print(f'D_t2_incorrect={D_t2_incorrect}')
````

In this assignment, we use the notations as given in the lecture notes. You may calculate the result with the help of a calculator or computer.

 

A set of training data (X) and their classes (Y) as shown below.

Find the highest normalized weight D just after calculation of step t =2.

**Anser:0.292**

````python
import math
import numpy as np
````

````python
# prepare the data
X = np.array([-15,37,2,54,32,4,42,10,22,39,-4,-33,-22,-25,-37,-31])
X = X.reshape(8,2)
Y = np.array([-1,-1,-1,-1,1,1,1,1])
size = len(Y)
D_old = np.array([1/size for i in range(size)])
````

````python
class BinaryClassifier:
    
    def __init__(self, polar, threshold):
        self.threshold = threshold
        self.polar = polar

    def predict(self, X, axis):
        predicts = []
        input_x = [X[i][axis] for i in range(len(X))]
        # debug printer
        # print(f"predict:{input_x} with thredhold:{self.threshold} and polar:{self.polar}")
        for x in input_x:
            if self.polar * x <= self.polar * self.threshold:
                predicts.append(1)
            else:
                predicts.append(-1)
        return predicts

    def calculateError(self, D_old, X, axis, true_labels):
        if len(D_old) != len(X):
            print("index error!")
        predicts = self.predict(X, axis)
        error = 0
        wrong = 0
        for i in range(len(predicts)):
            error += D_old[i]*1 if (predicts[i]!=true_labels[i]) else 0
            wrong += 1 if (predicts[i]!=true_labels[i]) else 0
        # print(f"made {wrong} mistakes")
        return error

# X should has only 2 demion
'''
    input:
        D_old: old D(bias) array
        X: input dataset
        true_labels: binary lables [-1,1]
        attr_index: axis of dataset need to consider
        polars: polar of classification hyper plane
    output:
        min_error: minimum error we found
        min_error_axis: the axis of selected attribute
        min_error_t: shreshold in selected attribute
'''
def searchBestClassifier(D_old, X, true_labels, 
                             attr_index=[0], polars=[1,-1]):
    data_size = len(X) 
    min_error = float("inf")
    min_error_axis = -1
    min_error_t = float("inf")
    min_error_polar = 1
    
    print(f"true labels:{true_labels}")
    
    for p in polars:
        # travel in thresholds
        attr_num = len(X[0])
        for axis in range(attr_num):
            errors = []
            # skip spliting at first element
            for t in X[:-1]:
                classifier = BinaryClassifier(polar=p, threshold=t[axis])
                error = classifier.calculateError(D_old=D_old, X=X, axis=axis, true_labels=true_labels)
                errors.append(error)
                # update and record
                if error < min_error:
                    min_error = error
                    min_error_axis = axis
                    min_error_t = t[axis]
                    min_error_polar = p
                    
            print(f"errors={errors} axis={axis} polar={p}")
            
    return min_error, min_error_axis, min_error_t, min_error_polar
    
# test
# searchBestClassifier(D_old, X, Y, [0,1])
````

````python
def calculate_D_new(D_old, X, true_labels):
    axis_space = [0,1]
    min_error, min_error_axis, min_error_t, min_error_polar = searchBestClassifier(D_old, X, true_labels, axis_space)
    print(f">>split at axis:{min_error_axis} threshold:{min_error_t} polar:{min_error_polar} error:{min_error}")
    alpha = 0.5 * math.log((1-min_error)/min_error)
    classifier = BinaryClassifier(polar=min_error_polar, threshold=min_error_t)
    pred_y = classifier.predict(X, min_error_axis)
    Z_t = 0
    D_new = np.zeros(size)
    for i in range(size):
        res = D_old[i] * math.pow(math.e, -1 * alpha * true_labels[i] * pred_y[i])
        Z_t += res
        D_new[i] = res
    # show right and wrong
    judge_strings = []
    for i in range(size):
        if true_labels[i] == pred_y[i]:
            judge_strings.append("✓")
        else:
            judge_strings.append("✘")
    print("predict:", judge_strings)
    # normalization with Z_t
    for i in range(len(D_new)):
        D_new[i] = D_new[i] / Z_t
    # print(f"D_new:{D_new}")
    return D_new
      
# test
# calculate_D_new(D_old, X, Y)
````

**QUESTION 14**

 In an object detection system using a detection window with size width=60 pixels, and height = 40 pixels. There are 3 basic feature types as shown below. The feature value is equal to (Sum of pixels in shaded area)- (Sum of pixels in white area). The numbers of different-scale features found in the detection window for type I, II, III are N1,N2,N3 respectively. Find N1+N2+N3.



**Answer:1457800**

````python
def count_features(size_x, size_y, split_x, split_y):
    max_width = int(size_x/split_x)
    max_height = int(size_y/split_y)
    res = 0
    for nx in range(1, max_width+1):
        for ny in range(1, max_height+1):
            num_x_blocks = size_x - nx*split_x + 1
            num_y_blocks = size_y - ny*split_y + 1
            res += num_x_blocks * num_y_blocks
    return res
# test with examples in ppt
# count_features(5, 5, 2, 1)
# count_features(24, 24, 3, 1)
# count_features(24, 24, 2, 2)

N1 = count_features(60, 40, 2, 1)
N2 = count_features(60, 40, 3, 1)
N3 = count_features(60, 40, 3, 2)

print(f"{N1}+{N2}+{N3}={N1+N2+N3}")
````
